В игре **Plants & Gardens** каждый месяц проводятся тематические события, ограниченные по времени. Игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время.

## Метрики для оценки результатов события

Для анализа успешности прошедшего события можно использовать следующие метрики:

1. **Количество завершенных уровней**
   - Общее количество уровней, пройденных игроком в рамках события.
   - Позволяет оценить, насколько эффективно игроки справляются с поставленными задачами. Например, если из 10 уровней игрок прошел 8, это указывает на успешное выполнение задания.
   

2. **Время выполнения**
   - Среднее время, затраченное игроком на прохождение каждого уровня.
   - Помогает понять, насколько сложно игрокам проходить уровни. Если среднее время превышает ожидаемое, это может указывать на проблемы с балансом уровней.
   

3. **Процент успешных попыток**
   - Доля уровней, которые были успешно пройдены с первой попытки.
   - Высокий процент (например, 80% и выше) может указывать на то, что уровни хорошо сбалансированы и не слишком сложны.
   

4. **Количество собранных предметов**
   - Уникальные предметы, полученные игроком в ходе события.
   - Эта метрика позволяет оценить привлекательность наград. Например, если игрок собрал 5 уникальных предметов, это может свидетельствовать о высокой заинтересованности в наградах.
   

5. **Количество использованных бонусов**
   - Общее количество бонусов, примененных игроком.
   - Помогает понять, насколько игроки активно используют доступные ресурсы для облегчения прохождения уровней. Например, если игрок использовал 10 бонусов, это может свидетельствовать о том, что уровни требовали дополнительной помощи.
   

6. **Монеты**
   - Количество дополнительных монет, полученных в ходе события.
   - Это также может быть индикатором успеха и прогресса игроков. Например, если игрок заработал 500 монет, это может указывать на хорошее выполнение заданий.

## Изменения метрик при усложнении механики

Если в другом событии усложнить механику так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад, набор метрик может измениться:

1. **Количество откатов**
   - Сколько раз игрок откатывался назад во время прохождения.
   - Эта метрика помогает оценить сложность уровней. Например, если игрок откатился 3 раза, это может указывать на необходимость корректировки сложности.
   

2. **Время на откат**
   - Время, затраченное игроком на возврат и повторное прохождение уровней.
   - Позволяет понять, насколько сильно сложность влияет на игровой процесс. Если время на откат превышает 5 минут, это может негативно сказаться на опыте игроков.
   

3. **Процент неудачных попыток**
   - Доля уровней, которые игрок не смог пройти с первой попытки.
   - Высокий процент (например, более 30%) может свидетельствовать о слишком высокой сложности уровней.
   

4. **Общее время на завершение**
   - Общее время, затраченное игроком на завершение события, включая время на откаты.
   - Эта метрика поможет получить полное представление о времени, проведенном игроком в игре. Например, если общее время составляет 40 минут, это может быть приемлемым значением.
   

5. **Уровень стресса**
   - Оценка, основанная на том, как игрок реагирует на откаты, например, количество попыток, которые закончились неудачей подряд.
   - Может быть измерен через опросы или аналитические данные, что поможет понять эмоциональное состояние игроков. Например, если 70% игроков сообщили о чувстве фрустрации, это сигнал о необходимости пересмотра сложности уровней.

## Пример

### Событие: "Летний фестиваль"

**Параметры события:**
- Длительность: 1 месяц
- Количество уровней: 10
- Награды: 5 уникальных предметов и 1000 монет

### Результаты:
- **Количество завершенных уровней:** 8 из 10
- **Время выполнения:** Среднее 4 минуты на уровень
- **Процент успешных попыток:** 85%
- **Количество собранных предметов:** 5 уникальных предметов
- **Количество использованных бонусов:** 10
- **Монеты:** 700 монет

### В случае усложнения механики:
**Параметры события:**
- При неудаче откат на 2 уровня назад

### Измененные результаты:
- **Количество откатов:** 4
- **Время на откат:** 10 минут
- **Процент неудачных попыток:** 35%
- **Общее время на завершение:** 60 минут
- **Уровень стресса:** 60% игроков сообщили о чувстве фрустрации

### Заключение

Таким образом, изменение механики событий требует пересмотра набора метрик для оценки результатов. Это позволяет глубже понять взаимодействие игроков с новыми механиками и их влияние на общий игровой опыт.
